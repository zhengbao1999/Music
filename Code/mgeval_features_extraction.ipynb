```python
import os
from numpy import array
o = r'D:\UNC\2021 Spring\Musical Similarity\continuation\selected_pairs_numbered_original' 
#dir to original (numbered)
c = r'D:\UNC\2021 Spring\Musical Similarity\continuation\selected_pairs_numbered_continuation'
#dir to continuation (numbered)

total_features = []
for i in range(len(os.listdir(o))):
    print(i)
    num_samples = 1
    set1 = [os.path.join(o, str(i) + '.mid')]
    set2 = [os.path.join(c, str(i) + '_continuation.mid')]
    set1_eval = {'total_used_pitch':np.zeros((num_samples,1)), 
             'total_used_note':np.zeros((num_samples,1)),
             #'total_pitch_class_histogram':np.zeros((num_samples,12)),
             #'pitch_class_transition_matrix':np.zeros((num_samples,12, 12)),
             'pitch_range':np.zeros((num_samples,1)),
             'avg_pitch_shift':np.zeros((num_samples,1)),
             'avg_IOI':np.zeros((num_samples,1))}
             #'note_length_hist':np.zeros((num_samples,12))}
             #'note_length_transition_matrix':np.zeros((num_samples,12,12))}

            # 这里是所有的feature
            # 两个1x12 两个12x12 都comment掉了
            
    set2_eval = {'total_used_pitch':np.zeros((num_samples,1)), 
             'total_used_note':np.zeros((num_samples,1)),
             #'total_pitch_class_histogram':np.zeros((num_samples,12)),
             #'pitch_class_transition_matrix':np.zeros((num_samples,12, 12)),
             'pitch_range':np.zeros((num_samples,1)),
             'avg_pitch_shift':np.zeros((num_samples,1)),
             'avg_IOI':np.zeros((num_samples,1))}
             #'note_length_hist':np.zeros((num_samples,12))}
             #'note_length_transition_matrix':np.zeros((num_samples,12,12))}

    metrics_list = list(set1_eval.keys())
    #print(metrics_list)
    for m in range(0, len(metrics_list)):
        #print(metrics_list[m])
        for i in range(0, num_samples):
            feature = core.extract_feature(set1[i])
            if not len(feature["pretty_midi"].instruments) < 1:
                set1_eval[metrics_list[m]][i] = getattr(core.metrics(), metrics_list[m])(feature)
    for m in range(0, len(metrics_list)):
        #print(metrics_list[m])
        for i in range(0, num_samples):
            feature = core.extract_feature(set2[i])
            if not len(feature["pretty_midi"].instruments) < 1:
                set2_eval[metrics_list[m]][i] = getattr(core.metrics(), metrics_list[m])(feature)
    #print(set1_eval)
    
    set1_features = []
    for i in set1_eval.keys():
        #print(list(set1_eval[i]))
        for j in set1_eval[i].tolist():
            for h in j:
                if type(h) == list:
                    for i in h:
                        set1_features.append(i)
                else:
                    set1_features.append(h)
    set2_features = []
    for i in set2_eval.keys():
        #print(list(set1_eval[i]))
        for j in set2_eval[i].tolist():
            for h in j:
                if type(h) == list:
                    for i in h:
                        set2_features.append(i)
                else:
                    set2_features.append(h)
    total_features.append(set1_features)
    total_features.append(set2_features)

#total_features里第0，1为第1组，以此类推2&3为第2组... 一共248组
print(array(total_features).shape)
```
