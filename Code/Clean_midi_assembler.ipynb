```python
import os
from os import path
import shutil

Root_Directory = r'D:\UNC\2021 Spring\Musical Similarity\Lahk\clean_midi'
Destination = r'D:\UNC\2021 Spring\Musical Similarity\Lahk\clean_midi_with_artist'
#dst_folder = os.mkdir(Destination)

d = {}
with open("file.txt") as f:
    for line in f:
       (key, val) = line.split()
       d[int(key)] = val

def main():
    for subdir, dirs, files in os.walk(Root_Directory):
        for dir in dirs:
            move_each_folder(dir)
            path = r'{}\{}'.format(Root_Directory, dir)
            if len(os.listdir(path)) == 0: # Check if the folder is empty
                shutil.rmtree(path)

def move_each_folder(foldername):
    Source_Path = r'{}\{}'.format(Root_Directory, foldername)
    #print(Source_Path)
    for filename in enumerate(os.listdir(Source_Path)):
        dst = foldername + "_-_" + filename[1]
        # rename all the files
        os.rename(os.path.join(Source_Path, filename[1]),  os.path.join(Destination, dst))


# Driver Code
if __name__ == '__main__':
    main()
```
